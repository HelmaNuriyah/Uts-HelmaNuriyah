```python
import matplotlib.pyplot as plt

# Data tahun
tahun = [2023, 2024, 2025, 2026, 2027, 2028]

# Data persentase untuk setiap status kualitas air
sangat_baik = [35, 32, 28, 25, 22, 20]
baik = [30, 28, 26, 25, 25, 25]
sedang = [20, 22, 24, 26, 28, 30]
buruk = [10, 12, 14, 16, 18, 20]
sangat_buruk = [5, 6, 8, 10, 12, 15]

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(tahun, sangat_baik, marker='o', label='Sangat Baik')
plt.plot(tahun, baik, marker='o', label='Baik')
plt.plot(tahun, sedang, marker='o', label='Sedang')
plt.plot(tahun, buruk, marker='o', label='Buruk')
plt.plot(tahun, sangat_buruk, marker='o', label='Sangat Buruk')

# Menambahkan judul dan label
plt.title('Prediksi Perubahan Kualitas Air Sungai Citarum (2023-2028)', fontsize=14)
plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Persentase (%)', fontsize=12)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.6)

# Menambahkan legenda
plt.legend(title="Status Kualitas Air", fontsize=10, title_fontsize=12)

# Menampilkan plot
plt.show()
```
